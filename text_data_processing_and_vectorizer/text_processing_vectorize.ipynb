```python
sentence_2 = "Yesterday morning, Professor Hasan Ali from Dhaka University met Sarah Johnson at the Google Bangladesh office to discuss the launch of the new Pixel X2 smartphone.  They talked about organizing a global AI Summit 2025 event in Tokyo next December.  Sarah mentioned that Sony Corporation plans to release the PlayStation Vita-3, while Microsoft will invest $500 million in cloud gaming.  Later that day, Hasan visited the National Art Museum to see the famous painting 'The Silent River'.  He also bought two books published by Penguin Random House and spoke briefly in Japanese with a tourist from Kyoto.  Construction work continues at the Padma Bridge facility, and the government announced that by 2027, all citizens will have access to 5G internet services."

tokens = word_tokenize(sentence_2)
pos_tags = pos_tag(tokens)
ner_tags = ne_chunk(pos_tags)
print(ner_tags)
```

```python
name = []
gpe = []
org = []

for chunk in ner_tags:
    if hasattr(chunk, 'label'):
        label = chunk.label()
        for word, tag in chunk:
            if label == 'PERSON':
                name.append(word)
            elif label == 'GPE':
                gpe.append(word)
            elif label == 'ORGANIZATION':
                org.append(word)

print(name)
print(org)
print(gpe)
```

```python
def prepocess(text):
    clean_words = ""
    stop_word_filter = ""
    processed = []

    for char in text:
        if char not in string.punctuation:
            clean_words = clean_words + char

    split_word = clean_words.split()

    for word in split_word:
        if word.lower() not in stop_word:
            processed.append(word)

    final_word = " ".join(processed)
    
    return final_word
```
