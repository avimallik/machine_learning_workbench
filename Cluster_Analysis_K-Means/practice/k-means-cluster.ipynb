```python
X = df.iloc[:, 1:3].values
# X
wcss = []
for i in range(1, 15):
    km1 = KMeans(n_clusters=i)
    km1.fit_predict(X)
    wcss.append(km1.inertia_)
wcss
```

```python
print(len(wcss))
```

```python
plt.plot(range(0, 14), wcss)
```

```python
kne = KneeLocator(range(0, 14), wcss, curve='convex', direction='decreasing')
```

```python
kne
```

```python
kne.plot_knee()
```
