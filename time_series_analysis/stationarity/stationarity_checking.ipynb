```python
import numpy as np
import pandas as pd
from statsmodels.tsa.stattools import adfuller
```

```python
pValue_const = 0.05
```

```python
temparature_ts_variant = [20, 22, 20, 21, 23, 21]
temparature_ts_variant_two = [50, 10, 30, 90, 70, 150, 110]
```

```python
result_ts_variant = adfuller(temparature_ts_variant)
result_ts_variant_two = adfuller(temparature_ts_variant_two)
```

```python
print(f'ADF Statistc : {result_ts_variant[0]}')
print(f'p-Value : {result_ts_variant[1]}')
print(f'Critical Valuses : {result_ts_variant[4]}')
```

```python
print(f'ADF Statistc : {result_ts_variant_two[0]}')
print(f'p-Value : {result_ts_variant_two[1]}')
print(f'Critical Valuses : {result_ts_variant_two[4]}')
```

```python
if result_ts_variant[1] <= pValue_const:
    print('Series is Stationary')
else:
    print('Series is Non-Stationary')
```

```python
if result_ts_variant_two[1] <= pValue_const:
    print('Series is Stationary')
else:
    print('Series is Non-Stationary')
```
